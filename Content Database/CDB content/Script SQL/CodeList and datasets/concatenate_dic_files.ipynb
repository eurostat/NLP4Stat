### Gather all the dictionnary name, code, label for each dictionnary in order to fill the mod_dictionnary and dat_code_dictionnary tables

```python
import pandas as pd
import os
```

```python
os.listdir(directory)
```

```python
test = {"code":[], "label":[], 'dico':[]}
directory = os.getcwd() ## can be replace by the path of the dictionnaries folder 
for filename in os.listdir(directory):
    if filename.endswith('.dic'):
        
        print(filename)
        file = open(filename, 'rb').read()
        file_arranged = file.decode('utf-8')
        file_arranged = file_arranged.split('\n')

        for row in file_arranged:
            if len(row.split('\t')) > 1:
                test['code'].append(row.split('\t')[0])
                test['label'].append(row.split('\t')[1].replace('\r', ''))
                test['dico'].append(filename.replace('.dic', ''))           
    else:
        continue
```

```python
df = pd.DataFrame.from_dict(test, orient = 'index')
df_final = df.T
df_final.to_excel('concatenate_dic.xlsx') 
```
